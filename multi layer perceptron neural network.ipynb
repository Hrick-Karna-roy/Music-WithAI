```python
import numpy as np

class MLN():
    
    def __init__(self,num_inputs = 3 , num_hidden_layer = [4 , 3] , num_outputs = 2):
        # constructing the attributes
        self.num_inputs = num_inputs
        
        self.num_hidden_layer = num_hidden_layer
        
        self.num_outputs = num_outputs
        
        self.layers = [num_inputs] + num_hidden_layer + [num_outputs]
        
        self.weights = []
        
        for i in range(len(self.layers)-1):
            w = np.random.rand(self.layers[i],self.layers[i+1])
            self.weights.append(w)
    
    
    # forward propagation method 
    def forword_propagation(self, inputs):
        # EACH INPUT LAYER IS ACTIVATION OF THE PREVIOUS LAYER
        
        activation = inputs

        for w in self.weights:
            # calculate net input by matrix multiply

            before_activation_net_input = activation@w

            # applying activation to net input

            activation = self._sigmoid(before_activation_net_input)

        return activation

    def _sigmoid(self,net_input_before_activation):
        return 1/(1+ np.exp(-net_input_before_activation))

n = MLN()
n.layers
n.weights
n.forword_propagation([.1,.9,2.3]) # INPUTS
```
